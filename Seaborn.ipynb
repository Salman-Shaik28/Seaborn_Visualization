```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/BAC.csv")
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month_name()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df.columns[1], y=df.columns[6], data=df)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Year', y='Close', data=df, estimator='mean')
plt.xlabel('Year')
plt.ylabel('Average Close Price')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df[df.columns[1]])
plt.title('Histogram Plot')
plt.show()
```

```python
sns.displot(df[df.columns[1]], kind='kde')
plt.title('Displot')
plt.show()
```

```python
sns.jointplot(x='High', y='Month', data=df, kind='scatter')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['Open', 'High', 'Low', 'Close', 'Adj Close']])
plt.ylabel('Price')
plt.show()
```

```python
sns.catplot(x='Month', y='Close', data=df, kind='violin', height=6, aspect=2)
plt.show()
```
